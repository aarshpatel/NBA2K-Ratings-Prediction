```python
def combine_stats(per_game_stats, advanced_stats, ratings, output_file):
    per_game = pd.read_csv(per_game_stats)
    advanced_stats = pd.read_csv(advanced_stats)
    only_advanced_stats = advanced_stats[advanced_stats.columns.difference(['Player', 'Age', 'Rk', 'G', 'MP'])]
    all_stats = pd.concat([per_game, only_advanced_stats], axis=1)
    all_stats["Player"] = map(lambda x: x.split("\\")[0], all_stats["Player"])
    all_stats = all_stats.drop("Unnamed: 19", axis=1)
    all_stats = all_stats.drop("Unnamed: 24", axis=1)
#     print(all_stats.columns)
    rk_ratings = pd.read_csv(ratings)
    subset_of_rk_ratings =  rk_ratings[["Player", "Team", "Rank"]]
#     subset_of_rk_ratings["Player"] = subset_of_rk_ratings["first_name"] + " " + subset_of_rk_ratings["last_name"]
#     subset_of_rk_ratings = subset_of_rk_ratings.drop("first_name", axis=1)
#     subset_of_rk_ratings = subset_of_rk_ratings.drop("last_name", axis=1)
    subset_of_rk_ratings.columns = ["Player", "Tm", "rating"]
    print subset_of_rk_ratings.head()
    all_stats = all_stats.loc[:,~all_stats.columns.duplicated()]
    combined_stats_with_ratings = pd.merge(all_stats, subset_of_rk_ratings, on=["Player", "Tm"])
    combined_stats_with_ratings = combined_stats_with_ratings.drop("Rk", axis=1)
    print combined_stats_with_ratings.head()
#     print(combined_stats_with_ratings.columns)
    new_columns = ["Player", "Position", "Age", "Team", "Games", "Games Started", "Minutes Player", "Field Goals Per Game", "Field Goal Attempts per Game", "Field Goal Percentage", "3-point field goals per game", "3-point field goal attempted per game", "3-point goal percentage", "2-point field goals per game", "2-point field goals attempted per game", "2-point field goal percentage", "effective field goal percentage", "free throws per game", "free throw attempts per game", "free throw percentage", "offensive rebounds per game", "defensive rebounds per game", "total rebounds per game", "assists per game", "steals per game", "blocks per game", "turnovers per game", "personal fouls per game", "points per game", "3 point attempt rate", "assist percentage", "block percentage", "box plus minus", "defensive box plus minux", "defensive rebound percentage", "defensive win shares", "free throw attempt rate", "offensive box plus minux", "offensive rebound percentage", "offensive win shares", "player efficiency rating", "steal percentage", "turnover percentage", "total rebound percentage", "true shooting percentage", "usage percentage", "value over replacement player", "win shares", "win shares per 48 minutes", "rating"  ]
    combined_stats_with_ratings.columns = [col.lower() for col in new_columns]
    combined_stats_with_ratings.to_csv(output_file, index=False)
```

```python
combine_stats("2001-per-game.csv", "2001-advanced-stats.csv", "raw_data/2001_player_data.csv", "preprocessed_data/2001_player_data.csv")
```
