```python
import pandas as pd
import numpy as np
```

```python
classifiers = ["Baseline Model", "Linear Regression", "Ridge Regression", "Random Forest Regressor"]
```

```python
optimized_scores = [6.28, 3.559, 3.534, 3.569]
default_scores = [6.28, 3.61, 3.607, 3.8655]
```

```python
df = pd.DataFrame({
        'Regression Model': classifiers,
        'Default Scores': default_scores,
        'Optimized Model Scores': optimized_scores
    })

df.to_csv("../data/scrape_data/default_vs_optimized.csv", index=False)
```

```python
import matplotlib
matplotlib.style.use('fivethirtyeight')
import matplotlib.pyplot as plt
%matplotlib inline

ax = df.plot.barh(x="Regression Model", figsize=(10,5))

ax.set_ylabel("MAE")
ax.set_title("Default Model vs Optimized Model", y=1.08)
plt.tight_layout()
fig = ax.get_figure()
fig.savefig("../graphs/default_mae_vs_optimized_model.png", dpi=200)
```
